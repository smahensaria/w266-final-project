```python
import json
import pandas as pd

from matplotlib import pyplot as plt
import seaborn as sb

import numpy as np
import math #needed for definition of pi

import statistics

import os

os.getcwd()
```

```python
def show_analysis(exploration_type, filename_BL, filename_FT, BL_string, FT_string):

    with open (filename_BL) as f_BL:
        data_BL = json.load(f_BL)

    with open (filename_FT) as f_FT:
        data_FT = json.load(f_FT)
        
    df_BL = pd.DataFrame.from_dict(data_BL)
    df_FT = pd.DataFrame.from_dict(data_FT)
    
    model_type_BL = BL_string
    meteor_scores_BL = []
    for m in data_BL:
        if model_type_BL in m:
            #print(m[model_type_BL]["scores"])
            if m[model_type_BL]["scores"]["meteor"] != -100:
                meteor_scores_BL.append(m[model_type_BL]["scores"]["meteor"])
            
    model_type_FT = FT_string
    meteor_scores_FT = []
    for m in data_FT:
        if model_type_FT in m:
            #print(m[model_type_FT]["scores"])
            if m[model_type_FT]["scores"]["meteor"] != -100:
                meteor_scores_FT.append(m[model_type_FT]["scores"]["meteor"])     
    
    #remove -100 meteor scores
    #if -100 in meteor_scores_BL: meteor_scores_BL.remove(-100)
    #if -100 in meteor_scores_FT: meteor_scores_FT.remove(-100)
    
    print(pd.DataFrame(meteor_scores_BL).describe())
    print(pd.DataFrame(meteor_scores_FT).describe())
        
    #plot frequency
    plt.plot(meteor_scores_BL, label="Baseline")
    plt.plot(meteor_scores_FT, label="Fine Tuned")
    plt.xlabel("Example")
    plt.ylabel("METEOR Score")
    plt.title(exploration_type + " METEOR Scores - Baseline vs Fine Tuned")
    plt.legend()
    plt.show()
    
 
    #plot histograms
    plt.hist(meteor_scores_BL, label="Baseline")
    plt.hist(meteor_scores_FT, label="Fine Tuned")
    plt.vlines(x=statistics.mean(meteor_scores_BL), ymin=0, ymax=max([1,len(meteor_scores_BL)/2]), colors='purple', linestyles="dashed")
    plt.text(statistics.mean(meteor_scores_BL), len(meteor_scores_BL)/2, round(statistics.mean(meteor_scores_BL), 2))
    plt.vlines(x=statistics.mean(meteor_scores_FT), ymin=0, ymax=max([1,len(meteor_scores_FT)/2]), colors='black', linestyles="dashed")
    plt.text(statistics.mean(meteor_scores_FT), len(meteor_scores_FT)/2, round(statistics.mean(meteor_scores_FT), 2))
    plt.title(exploration_type + " METEOR Scores - Baseline vs Fine Tuned")
    plt.legend()
    plt.show()
    
    #plot distributions
    sb.distplot(meteor_scores_BL, hist=False, label="Baseline")
    sb.distplot(meteor_scores_FT, hist=False, label="Fine Tuned")
    #sb.displot(meteor_scores_BL, kind="kde", label="Baseline")
    #sb.displot(meteor_scores_FT, kind="kde", label="Fine Tuned")
    
    plt.vlines(x=statistics.mean(meteor_scores_BL), ymin=0, ymax=max([1,12]), colors='purple', linestyles="dashed")
    plt.text(statistics.mean(meteor_scores_BL), 12, round(statistics.mean(meteor_scores_BL), 2))
    plt.vlines(x=statistics.mean(meteor_scores_FT), ymin=0, ymax=max([1,12]), colors='black', linestyles="dashed")
    plt.text(statistics.mean(meteor_scores_FT), 12, round(statistics.mean(meteor_scores_FT), 2))
    plt.xlabel("METEOR Score")
    plt.title(exploration_type + " METEOR Scores - Baseline vs Fine Tuned")
    plt.legend()
    plt.show()
    
    return meteor_scores_BL, meteor_scores_FT
```

```python
exploration_type = "KP Sampling"
filename_BL = ".\\data\\baseline_kpsampling_results.json"
filename_FT = ".\\data\\kpsampling_results.json"

BL_string = "base_sampling"
FT_string = "sampling"

meteor_KPsampling_BL= []
meteor_KPsampling_FT = []
meteor_KPsampling_BL, meteor_KPsampling_FT = show_analysis(exploration_type, filename_BL, filename_FT, BL_string, FT_string)
```
